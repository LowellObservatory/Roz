# Hull Finder
This notebook contains the hull finding and improving methods. A "hull" is an outline of points. There's no real algorithm for finding a non-convex hull, so this notebook implements a rough method to do so. Saves the hull to a text file at the end. This notebook should run cleanly all the way through with graphics to ensure it's been done right.

The following two cells replot the DESI tiles in an Eckert-IV projection.

```python
# Newton's method for eckert-iv proection.
def eckertiv_findtheta(phi, n):
    # First short circuit
    if n == 0:
        return phi/2

    pi = math.pi

    # Array literally just filled with half pis.
    halfpi = np.empty(len(phi))
    halfpi.fill(pi/2)

    theta = eckertiv_findtheta(phi, n-1)

    # Condition for the angle is pi/2 OR -pi/2
    cond1 = np.equal(theta, halfpi)
    cond2 = np.equal(theta, -1*halfpi)
    cond = np.logical_or(cond1, cond2)

    # Choose the original value (pi/2 or -pi/2) if its true for equality
    # Otherwise use that value's thetanew.
    # This is the eckertiv theta finding Newton's method.
    # It's been broken up for style.
    s_theta = np.sin(theta)
    c_theta = np.cos(theta)
    num = theta + np.multiply(s_theta, c_theta) + 2 * s_theta - (2 + pi/2) * np.sin(phi)
    denom = 2 * c_theta * (1 + c_theta)
    thetanew = theta - num/denom
    thetanew = np.where(cond, phi, thetanew)

    return thetanew


def eckertiv(ra, dec):
    # Center latitude
    center = math.radians(180)

    # n = 5 seems to be sufficient for the shape.
    # This doesn't converge as quickly as Mollweide
    theta = eckertiv_findtheta(np.radians(dec), 5)

    R = 100

    # For readability sake
    coeff = 1/math.sqrt(math.pi*(4+math.pi))

    # Eckert IV conversion functions.
    x = 2 * R * coeff * np.subtract(np.radians(ra), center) * (1 + np.cos(theta))
    y = 2 * R * math.pi * coeff * np.sin(theta)

    return(x, y)
```

First we split the DESI tiles into two sections. This at first is just a test of the convex hull fiding method. I implemented the Jarvis method (also called the gift wrapping method) for finding a convex hull. Wikipedia has a pretty decent explanation of this method: https://en.wikipedia.org/wiki/Gift_wrapping_algorithm. The detriments of the convex hull can be seen in the plot, where the lower riht corner doesn't wrap around the inner gaps of the tiles.

```python
mid = []
right = []
left = []

tiles = astropy.table.Table.read('desi-tiles.fits', hdu=1)
sel = (tiles['IN_DESI'] > 0) & (tiles['PASS'] < 8)
ra = tiles['RA'][sel]
dec = tiles['DEC'][sel]

for i in range(0,len(ra)):
    if 90 < ra[i] < 300:
        mid.append((ra[i], dec[i]))
    elif 300 < ra[i]:
        right.append((ra[i], dec[i]))
    else:
        left.append((ra[i], dec[i]))
```

```python
def jarvis(points):
    #print(points.type)
    points = list(set(points))
    sort = sorted(points)
    
    # Starting at the rightmost point rather than the left most has been more accurate.
    start = sort[-1]
    
    # The start has no "previous point" so just use a point slightly above the start.
    curr = nextpoint(start, (start[0], start[1] + 1), points)
    
    # The hull is kept in the result array.
    result = []
    result.append(start)
    result.append(curr)
    
    prev = start
    nextp = (0,0)
    
    # Loop that continues stepping along the hull until we're back at the start
    i = 0
    while not nextp == start:
        nextp = nextpoint(curr, prev, points)
        result.append(nextp)
        prev = curr
        curr = nextp
        i += 1
    
    return result


def nextpoint(curr, prev, points):
    
    thetalist = []
    potentials = []
    
    # Putting these outside the loop means we don't have to run it every time.
    v_p = (prev[0] - curr[0], prev[1] - curr[1])
    v_p = v_p / np.linalg.norm(v_p)
    
    for point in points:
        v_c = (point[0] - curr[0], point[1] - curr[1])
        v_c2 = v_c / np.linalg.norm(v_c)
        
        # Always finds the inner angle between the check point and the previous hull point.
        theta = np.arccos(np.clip(np.dot(v_c2, v_p), -1.0, 1.0))
        
        
        cross = np.cross(v_p, v_c)
        #if cross < 0:
            #theta = 2*math.pi - theta
        
        thetalist.append(theta)
        
    
    sort = sorted(thetalist)
    index = thetalist.index(sort[-1])
    
    #print(sort[-1])
    return points[index]

```

Next I experimented with a method to turn the convex hull slightly concave, especially in the lower right. First I bisected each sue and added the nearest point to the hull. Afterwards I would correct to ensure that no points accidentally ended up outside the hull. In the plot above the lower left corner there's a single point that ended up outside the hull. Correct() adds points like these to the hull itself. 

```python
# This method takes the hull, and uses the DESI tile points. It runs through the points on the hull 
# and bisects the line. It then finds the nearest point in the DESI tiles and adds it to the hull.
# Sides are only bisected if they're longer than length 5.
def hull_bisect(hull, points):

    newhull = []
    for i in range(0,len(hull)):
        curr = hull[i]
        if not i == len(hull) - 1:
            nextp = hull[i + 1]
        else:
            nextp = hull[0]
    
        midpoint = ((curr[0] + nextp[0]) / 2, (curr[1] + nextp[1]) / 2)
        v1 = (nextp[0] - curr[0], nextp[1] - curr[1])
        d1 = np.linalg.norm(v1)
        newhull.append(curr)
        
        # We only want to splice if the length of the side is longer than a certain length.
        if d1 > 5:
            #print(d1)
            l = 100
            best = (0,0)
    
            for point in points:
                v = (point[0] - midpoint[0], point[1] - midpoint[1])
                d = np.linalg.norm(v)
        
                if d < l:
                    l = d
                    best = point
        
            if not best == (0,0):
                newhull.append(best)
        
    return newhull
```

```python
def correct(hull, points):
    # This method ensures that any points outside the hull get added to the hull.
    # This is mainly in case the bisection took the hull inside the DESI tiles.
    new = list(hull)
    hullshape = path.Path(hull)
    
    for point in points:
        if not hullshape.contains_point(point):
            new.append(point)
      
    
    # Remove any duplicates we picked up on the way
    new = list(set(new))
    new = clockwise_sort(new)
    return new
        
def clockwise_sort(tosort):
    # Sorts points clockwise. I later ported this code over into transform.py.
    tosort = np.array(tosort)
    x = tosort[:,0]
    y = tosort[:,1]
    
    x = x.tolist()
    y = y.tolist()
    
    x = sorted(x)
    y = sorted(y)
    
    centerx = (x[0] + x[-1])/2 + 5
    centery = (y[0] + y[-1])/2 - 5
    
    center = (centerx, centery)
    
    tosort = np.subtract(tosort, center)
    
    # Slices out the x and y portions
    x = tosort[:,0]
    y = tosort[:,1]
    
    # Creates polar nonsense
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    
    # Reshape to stack
    r = np.reshape(r, (r.shape[0], 1))
    theta = np.reshape(theta, (theta.shape[0], 1))
    
    # Stack into list of form (theta,r) (we want to sort theta first)
    stack = np.hstack((theta,r))
    
    stack2 = []
    for i in stack:
        stack2.append(tuple(i))
        
    # Standard python sort by theta
    stack2 = sorted(stack2)
    
    # Now we just have to convert back!
    # Slice out theta and r
    stack2 = np.array(stack2)
    theta = stack2[:,0]
    r = stack2[:,1]
    
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    
    result = []
    for i in range(0,len(x)):
        result.append((x[i] + center[0],y[i] + center[1]))
    
    return result
    
```

```python
def angles(hull):
    # This method finds corners whose angles are too large. This helps to "smooth" over the
    # Hull, as corners that are too jagged would get removed.
    vertices = list(hull)
    
    thetalist = []
    remove = []
    for i in range(0,len(vertices)):
        curr = vertices[i]
        
        # Code for carryover at the end
        if not i == len(vertices) - 1:
            nextp = vertices[i + 1]
        else:
            nextp = vertices[0]
            
        if not i == 0:
            prev = vertices[i-1]
        else:
            prev = vertices[-1]
            
        # Finds the angle using dot product
        v_p = (prev[0] - curr[0], prev[1] - curr[1])
        v_p = v_p / np.linalg.norm(v_p)
        
        v_n = (nextp[0] - curr[0], nextp[1] - curr[1])
        v_n = v_n / np.linalg.norm(v_n)
        
        theta = np.arccos(np.clip(np.dot(v_n, v_p), -1.0, 1.0))
        
        # Make the angle the inner angle if we found the outer one
        if np.cross(v_n, v_p) <= 0:
            theta = 2*math.pi - theta
        if theta > 5.2:
            remove.append(curr)
        
        thetalist.append(theta)
    
    newhull = list(vertices)
    
    for point in remove:
        newhull.remove(point)
    
    newhull = clockwise_sort(newhull)
    return newhull
```

```python
def improve(hull, points):
    # Runs a hull bisect, the corrects it and runs the angles method.
    # Angles is run three times for maximum smoothing.
    temp = hull_bisect(hull, points)
    temp = correct(temp, points)
    temp = angles(temp)
    temp = angles(temp)
    return angles(temp)
```

Since the plot in the transform.py methods is a little small, we want the hull to be slightly larger. Since the hull was built using tiles in the DESI viewing area, the hull crosses directly over some of the points, and includes many of the tiles in the edge. Hence we expand the hull along the angle of each corner by 2 degrees. This way the hull includes the DESI area without running over it.

```python
def expand_hull(hull):
    expand = np.array(hull)
    x = expand[:,0]
    y = expand[:,1]
    
    x = x.tolist()
    y = y.tolist()
    
    x = sorted(x)
    y = sorted(y)
    
    centerx = (x[0] + x[-1])/2 + 5
    centery = (y[0] + y[-1])/2 - 5
    
    center = (centerx, centery)
    
    expand = np.subtract(expand, center)
    
    # Slices out the x and y portions
    x = expand[:,0]
    y = expand[:,1]
    
    # Creates polar nonsense
    # Then expands in r by expansion
    expansion = 2
    r = np.add(np.sqrt(x**2 + y**2),expansion)
    theta = np.arctan2(y, x)
    
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    
    result = []
    for i in range(0,len(x)):
        result.append((x[i] + center[0], y[i] + center[1]))
    
    #print(result)
    return result
    
```
